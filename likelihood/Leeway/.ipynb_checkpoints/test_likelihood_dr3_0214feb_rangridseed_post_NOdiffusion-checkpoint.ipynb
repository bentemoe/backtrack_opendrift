# Running backwards simulation: OpenDrift - Leeway model

First, we will compute a backwards run to have an indication for the seeding area for the forward run. 
A grid will be created based on this backward to do a uniform random seeding for the forward run with equal amount of seeded particles per grid cell. 

Import Eulerian ocean current and wind data
1. Ocean: Globcurrent
2. Wind: ECMWF

```python
reader_globcur = reader_netCDF_CF_generic.Reader("C:/Users/bente/Documents/THESIS/Globcurrent/*.nc")
reader_wind = reader_netCDF_CF_generic.Reader('C:/Users/bente/Documents/THESIS/Globcurrent/Simulations_24h/OO-2023-03/agulhas_wind_2023.nc',
        standard_name_mapping={'u10': 'x_wind',
                               'v10': 'y_wind'})

# Choosing the object type (with certain Leeway coefficients) to run the simulations with
object_type = 78  # 78 = Sewage floatables; downwind 1.8 percent wdf; no offset but stdev around 3 cm/s; right/left slope 0.16 percent
```

Choosing duration and start date of the simulations. Backward simulations should be initiated from end_time. 

```python
duration = timedelta(hours=288)                # 5 days
start_time = datetime(2023,2,2,3,12,8)
end_time = start_time + duration
```

```python
outfile_bwd = "D:/backtrack/oceandrift_48h/likelihood/lw_bwd0214feb_nodif100.nc"
```

Loading the model, adding readers, set configurations, define the seeding and run the model/simulations. 

```python
orev = Leeway(loglevel=50)                                         # Import Leeway model
orev.add_reader([reader_globcur, reader_wind])                     # Import the Eulerian ocean current and wind data into the model
orev.set_config('drift:advection_scheme', 'runge-kutta4')          # Set advection scheme to RK4 (default is Euler fwd; but especially for backwards simulations, higher order is better)
#orev.set_config('drift:horizontal_diffusivity', 20)               # Optional: turn on diffusivity 
orev.seed_elements(lon=startlon, lat=startlat, radius=10000, radius_type='uniform', number=100, time=end_time, object_type=object_type)      # Seeding from end time and end location in radius of 10km of observed drifter; number of 100 to get indication of the area to seed for forward run
orev.run(duration=duration, time_step=-900, time_step_output=3600, outfile=outfile_bwd)       # Run the model with NEGATIVE TIME STEP; Time step here is 10min: also here, especially for backward runs, the smaller the time step, the better
```

```python
back = xr.open_dataset(outfile_bwd)
```

### Creating grid based on backward run

```python
# Set the desired grid size
grid_size_lon = 0.480  # Change this to your desired grid size
grid_size_lat = 0.400

# Calculate the number of grid cells in each dimension
lon_bins = np.arange(np.min(back.isel(time=288).lon.values)-(0.5), np.max(back.isel(time=288).lon.values)+(0.5), grid_size_lon)
lat_bins = np.arange(np.min(back.isel(time=288).lat.values)-(0.5), np.max(back.isel(time=288).lat.values)+(0.5), grid_size_lat)
```

```python
binmesh = np.meshgrid(lon_bins, lat_bins)
```

### Create seeding locations for forward run: uniform random seeding with equal amount of particles per grid cell

```python
lons = np.zeros((len(lon_bins)-1,(len(lat_bins)*200)-200))
lats = np.zeros((len(lon_bins)-1,(len(lat_bins)*200)-200))
for i in range(len(binmesh[0][0])-1):
    for j in range(len(binmesh[1])-1):
        lons[i][(j*200):((j+1)*200)] = np.random.uniform(binmesh[0][0][i], binmesh[0][0][i+1], 200) # Much coarser CCC
        lats[i][(j*200):((j+1)*200)] = np.random.uniform(binmesh[1][j][0], binmesh[1][j+1][0], 200)
```

```python
outfile = "D:/backtrack/oceandrift_48h/likelihood/lw_fwd0214feb_nodif.nc"
```

```python
ofwd = Leeway(loglevel=50)
ofwd.add_reader([reader_globcur, reader_wind])
ofwd.set_config('drift:advection_scheme', 'runge-kutta4')
#ofwd.set_config('drift:horizontal_diffusivity', 20)
ofwd.seed_elements(lon=lons, lat=lats, radius=0, time=start_time, object_type=object_type)
ofwd.run(duration=duration, time_step=900, time_step_output=3600, outfile=outfile)
```

### Retrieve end locations of all forward runs to see which end up in a 10km radius of the 'end location' of the observed drifter. We call these trajectories 'hits'.

```python
## FORWARD ## 
lon0 = []
lat0 = []
lonend0 = []
latend0 = []
for j in range(len(fwd.trajectory)):
        ### calculating hits ###
        lon0.append(fwd.isel(trajectory=j).lon.values)          # All lon of simulated trajectory
        lat0.append(fwd.isel(trajectory=j).lat.values)          # All lat of simulated trajectory
        lonend0.append(fwd.isel(trajectory=j).lon.values[-1])   # lon at end of simulated trajectory
        latend0.append(fwd.isel(trajectory=j).lat.values[-1])   # lat at end of simulated trajectory
        
geod = pyproj.Geod(ellps='WGS84')
on = np.ones(len(lonend0))
dummy1, dummy2, dist2incident = geod.inv(lonend0, latend0, ilon*on, ilat*on)   # Calculate distance between end location of observed drifter and simulated particle
hits = np.where(dist2incident<10000)       # Check which trajectories make it into a 10km radius of end location of the observed drifter, we call these trajectories 'hits'

# Retrieve start locations of the 'hitting' simulated particles
hit_start_lons0= []
hit_start_lats0= []
for m in range(len(hits)):
        hit_start_lons0.append(lon0[hits[0][m]][0])
        hit_start_lats0.append(lat0[hits[0][m]][0])
```

# Analysis 'hitting' FWD simulated trajectories

In this section we will calculate the grid count ($n_j$), the likelihood, prior probability and posterior probabilty per grid cell. 

```python
fwd = xr.open_dataset(outfile)
```

```python
# Seperate dataset for the FWD 'hitting' trajectories
xfwd = fwd.isel(trajectory=hits[0])
```

#### Retrieve start lon and lat of FWD 'hitting' trajectories

```python
lon = []       # lon for 'hits'
lat = []       # lat for 'hits'

for i in range(len(xfwd.trajectory)):
    lon.append(xfwd.isel(trajectory=i).lon.values[0])
    lat.append(xfwd.isel(trajectory=i).lat.values[0])
```

#### Retrieve start lon and lat of ALL FWD simulated trajectories

```python
lonall = []
latall = []

for i in range(len(fwd.trajectory)):
    lonall.append(fwd.isel(trajectory=i).lon.values[0])
    latall.append(fwd.isel(trajectory=i).lat.values[0])
```

### Count number of starting point of FWD 'hittin' simulated trajectories per grid cell ($n_j$)

```python
# Function
def count_points_in_grid(lon, lat, lonbins, latbins):
    
    lobi = []
    labi = []
    for i in range(len(lonbins)-1):
        lobi.append((lonbins[i]+lonbins[i+1])/2)
    for i in range(len(latbins)-1):
        labi.append((latbins[i]+latbins[i+1])/2)
        
    # Count the number of points in each grid cell
    counts, xedges, yedges = np.histogram2d(lon, lat, bins=[lonbins, latbins])
    
    # Create a new DataArray with the counts and assign coordinates
    counts_dataarray = xr.DataArray(counts, coords=[('lon_bin', lobi), ('lat_bin', labi)])
    
    return counts_dataarray 

# Call the function to count points in grid cells for 'hitting' simulated trajectories
grid_counts = count_points_in_grid(lon, lat, lon_bins, lat_bins)
```

### Plotting 'hitting' trajectories with grid count ($n_j$)

```python
o_hit.plot(background=grid_counts, clabel='$n_j$', fast=True, markersize=.5, vmin=0, cmap='YlOrRd')
```

#### Count number of particles per grid cell ($n_j$) for all trajectories => SHOULD BE EQUAL NUMBER FOR EVERY GRID CELL !! 

```python
# Call the function to count points in grid cells
grid_counts_all = count_points_in_grid(lonall, latall, lon_bins, lat_bins)
```

```python
o_hit.plot(background=grid_counts_all, clabel='$N_j$', fast=True, markersize=.5, vmin=0, cmap='YlOrRd')  
```

# LIKELIHOOD 

```python
o_hit.plot(background=grid_counts/grid_counts_all, clabel='$n_j / N_j$', fast=True, markersize=.5, vmin=0, cmap='YlOrRd')
```

```python
likelihood = grid_counts/grid_counts_all
```
